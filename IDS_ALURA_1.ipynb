**Desafio 07:** Deixar os grÃ¡ficos bonitÃµes. (Matplotlib.pyplot)

```python
# Desafio 7

# Iremos utilizar a biblioteca 'MatplotLib' para alterar visualmente os graficos 

import matplotlib.pyplot as plt

plt.style.use('ggplot')

# Define o tamanho da figura
plt.figure(figsize=(6,6))

# Plota o grafico, "color" define a cor das barras, "ec" define a cor da borda das barras, "k" Ã© uma abreviaÃ§Ã£o para black
# Alpha define a opacidade das barras
dados['tratamento'].value_counts().plot.bar(color="teal", ec="k", alpha=0.7)

# Define um titulo para o grafico
plt.title('Total de tratamentos')

# Define uma label para o eixo x do grafico
plt.xlabel("Tipo de tratamento")

# Define uma label para o eixo y do grafico
plt.ylabel("NÃºmero de tratamentos efetivados")

```

```python
dose = dados['dose'].value_counts()
tratamentos = dados['tratamento'].value_counts(normalize = True)

plt.figure(figsize=(16, 6))

# Plota mais de um grafico na mesma figura (linhas, colunas, plot selecionado)
plt.subplot(1, 2, 1)

dose.plot.bar(ec="k", color="royalblue", alpha=.6)
plt.title('Doses aplicadas')
plt.xlabel('Doses aplicadas')
plt.ylabel('NÃºmero de doses')

plt.subplot(1, 2, 2)
tratamentos.plot.pie(labeldistance=1.1, explode=[0, .15], textprops={"fontsize": 12})

# Define uma margen para o subplot
plt.subplots_adjust(right = 1)

# Garante a forma circular do grafico
plt.axis("equal")

# Cria uma legenda para cada cor
plt.legend()

plt.title('ProporÃ§Ã£o de tratamentos')
plt.ylabel('')
```
